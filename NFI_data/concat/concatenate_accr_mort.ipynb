```python
import pandas as pd
```

```python
data_m_1 = pd.read_excel('./DATA_Français/NUTZUNG-MORTALITÄT_FRENCH.xlsx', sheet_name='LFI1-LFI2')
data_m_2 = pd.read_excel('./DATA_Français/NUTZUNG-MORTALITÄT_FRENCH.xlsx', sheet_name='LFI2-LFI3')
data_m_3 = pd.read_excel('./DATA_Français/NUTZUNG-MORTALITÄT_FRENCH.xlsx', sheet_name='LFI3-LFI4')
```

```python
data_m_3.head()
```

```python
data_m = pd.concat([data_m_1, data_m_2, data_m_3])
data_m.info()
```

```python
data_m.head()
```

```python
data_accr_1 = pd.read_excel('./DATA_Français/ZUWACHS_FRENCH.xlsx', sheet_name='LFI1-LFI2')
data_accr_2 = pd.read_excel('./DATA_Français/ZUWACHS_FRENCH.xlsx', sheet_name='LFI2-LFI3')
data_accr_3 = pd.read_excel('./DATA_Français/ZUWACHS_FRENCH.xlsx', sheet_name='LFI3-LFI4')
```

```python
data_accr = pd.concat([data_accr_1, data_accr_2, data_accr_3])
len(data_accr)
```

```python
data_join = data_m.merge(data_accr,on=['PARCELLE','LFI','X','Y','NB_SAISON'],how='outer')
data_join.head()
```
